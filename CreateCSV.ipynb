```python
from os import walk
import pandas as pd
import codecs
from codecs import open


df = pd.DataFrame(columns=['queryID','variantID','queryText'])
mypath = "./submissions/"
f = []
f2 = []
for (dirpath, dirnames, filenames) in walk(mypath):
    print("DIRNAMES: ",dirpath, "FNs: ", filenames)
    
    variantNum = dirpath[14:]
    for filename in filenames:
        if not filename.startswith('.'):
            fname = dirpath+"/"+filename
            sql = open(fname, mode='r', encoding='utf-8-sig').read()
            if sql.lower().startswith('select'):
#                 print(sql)
                new_row = {'queryID':filename[:-4], 'variantID':variantNum, 'queryText':sql}
                #append row to the dataframe
                df = df.append(new_row, ignore_index=True)
df.to_csv('fa19_queries.csv',index=False)
```
