### Read reserved words 

We want to pay attention to the ordering of these words, IF they are outside quotations

Source: https://www.drupal.org/docs/develop/coding-standards/list-of-sql-reserved-words

```python
keywords=set()
with open('sql_reserved_words.csv') as words:
    reader = csv.reader(words)
    for row in reader:
        keywords.add(row[0])
```

# 1. Retrieve Queries

```python
def qPrint(query):
    print(query[0] + "\t" + query[2] + "\t" + query[1])
```

```python
queries=[]
qCounter=0
with open('sql_queries.tsv') as tsv:
    reader = csv.reader(tsv, delimiter='\t')
    for idx, row in enumerate(reader):
        if idx!=0:
            # questionId, queryContent, queryId
            entry=[]
            entry.append(row[0])
            entry.append(row[1])
            entry.append(str(qCounter))
            qCounter+=1
            queries.append(entry)

print("quesId\tqueryId\tQueryContent")
for i in range(0,5):
    qPrint(queries[i])
        
```
